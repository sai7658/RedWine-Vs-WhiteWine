# white wine

+ve correaltion with quality:

1. Alcohol
2. pH(weak)

-ve correaltion with quality:

1. volatile Acidity
2. Total sulfur dioxide
3. density
4. chlorides
5. residual sugars(weak)

# combine the datasets

```python
red_wine_df['type'] = 'Red'
white_wine_df['type'] = 'White'
```

```python
wines_df = pd.concat([red_wine_df,white_wine_df])
```

```python
wines_df.info()
```

# comparitive Analysis

```python
sns.countplot(x='quality',hue='type',data = wines_df)
```

```python
plt.figure(figsize=(7,7))
sns.kdeplot(red_wine_df['quality'],shade=True,color='r',label='red wine')
sns.kdeplot(white_wine_df['quality'],shade=True,color='b',label='white wine')
```

```python
sns.boxplot(x='quality',y='alcohol',hue='type',data = wines_df,showfliers=False,palette=['r','w'])
```

```python
sns.boxplot(x='quality',y='density',hue='type',data = wines_df,showfliers=False,palette=['r','w'])
```

```python
sns.jointplot(x='alcohol',y='residual sugar',data = red_wine_df)
```

```python
sns.boxplot(x='quality',y='residual sugar',hue='type',data = wines_df,showfliers=False,palette=['r','w'])
```

Red wine has very less sugars compared to white wine

```python
plt.figure(figsize=(7,7))
sns.kdeplot(red_wine_df['residual sugar'],shade=True,color='r',label='red wine')
sns.kdeplot(white_wine_df['residual sugar'],shade=True,color='b',label='white wine')
```

```python
sns.regplot(x='alcohol',y='residual sugar',data = white_wine_df)
```
